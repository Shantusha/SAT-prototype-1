```python
import pandas as pd
import csv
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Adding the dataframe
df = pd.read_csv('nail dataset.csv', sep=';')
df.head
```

```python
# Checking the values ​​of the dataset
df.info()
```

### Clean data

```python
# Rename column names
df = df.rename(columns={"nail art level":"nailartLevel"})
df = df.rename(columns={"nail art style":"styles"})
df = df.rename(columns={"nail art ":"nailart"})

# split data of column styles
def split_values(row):
    values = row['styles'].split(',')
    return pd.Series(values)

new_df = df.apply(split_values, axis=1)
new_df.columns = ['style{}'.format(i+1) for i in range(new_df.shape[1])]
nail_df = pd.concat([df, new_df], axis=1)
nail_df[['style1', 'style2', 'style3', 'style4']] = nail_df[['style1', 'style2', 'style3', 'style4']].fillna("-")

nail_df

```

```python
# Presenting the questions that need to be asked to the user
# For this example I used the following input: color: blue, shape: round, length: long, nail art: yes, level: simple, style: french tips
inputColor = input("What color do you like to wear? (green, orange, blue, purple, yellow or red)")
inputShape = input("What shape do you like for your nails? (square, round or coffin)")
inputLength = input("What nail length do you prefer? (short, middle long or long)")
inputNailart = input("Do you want nail art on your nails? (yes/no)")
inputNailartLevel = input("What type of nailart would you like? (simple, medium or complex)")
inputNailartStyle = input("What nail art style do you like? (line art, swirl, patterns, dots, glitters, stones, small art, medium art, big art, french tips, reversed tip, hearts, flowers, checkered, stickers, ombre, marmer, golden flakes)")
```

```python
# Using the color input as a new dataframe to only calculate the similarities of the user input on this color.
# Inspiration code from Isha Bansal: https://www.askpython.com/python/examples/convert-number-to-words 
color_like = nail_df[nail_df['color'].str.contains(inputColor)]

# Give score for nail shape
def like_shape(input):
    nail_dataset = input
    
    if (nail_dataset == inputShape):
        return 1
    else:
        return 0
    
    
color_like["score_shape"] = color_like["shape"].apply(like_shape)
color_like.head(5)
```

```python
# Give score for length of the nail

def like_length(input):
    nail_dataset = input
    
    if (nail_dataset == inputLength):
        return 1
    else:
        return 0   
    
color_like["score_length"] = color_like["length"].apply(like_length)
color_like.head(5)

```

```python
# Give score for the use of nail art

def use_nailart(input):
    nail_dataset = input
    if (nail_dataset == inputNailart):
        return 1
    else:
        return 0   
    
color_like["score_nailartUse"] = color_like["nailart"].apply(use_nailart)
color_like.head(5)
```

```python
# Give score for nail art level
def nailart_level(input):
    nail_dataset = input
    if (nail_dataset == inputNailartLevel):
        return 1
    else:
        return 0   
    
color_like["score_nailarLevel"] = color_like["nailartLevel"].apply(nailart_level)
color_like.head(5)
```

```python
# give score for nailart style

def like_nailart(input):
    nail_dataset = input
    if (nail_dataset == inputNailartStyle):
        return 1
    else:
        return 0   
    
color_like["score_style1"] = color_like["style1"].apply(like_nailart)
color_like["score_style2"] = color_like["style2"].apply(like_nailart)
color_like["score_style3"] = color_like["style3"].apply(like_nailart)
color_like["score_style4"] = color_like["style4"].apply(like_nailart)
color_like
```

```python
# Add up the points and divide by the number of questions asked. Multiply the result of this by 100 to get a percentage

color_like["Final_score"] = (color_like["score_shape"] + color_like["score_length"] + color_like["score_nailartUse"] + color_like["score_nailarLevel"] + color_like["score_style1"] + color_like["score_style2"] + color_like["score_style3"] + color_like["score_style4"])
color_like["Final_score"] = (color_like["Final_score"] / 6) * 100
color_like = color_like.sort_values(by="Final_score", ascending=False)
color_like.head(5)
```

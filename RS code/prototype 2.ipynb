```python
import pandas as pd
import csv
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Adding the dataframe
df = pd.read_csv('nail dataset_prototype_2.csv', sep=';')
df.head
```

```python
# Checking the values ​​of the dataset
df.info()
```

```python
# Rename column names
df = df.rename(columns={"color group":"colorGroup"})
df = df.rename(columns={"nail art level":"nailartLevel"})
df = df.rename(columns={"nail art style":"styles"})
df = df.rename(columns={"nail art ":"nailart"})
df = df.rename(columns={"topic effect ":"topicEffect"})
#nail_df

# split data of column color
def split_values(row):
    values = row['color'].split(',')
    return pd.Series(values)

new_df = df.apply(split_values, axis=1)
new_df.columns = ['color{}'.format(i+1) for i in range(new_df.shape[1])]
df = pd.concat([df, new_df], axis=1)
df[['color1', 'color2', 'color3', 'color4']] = df[['color1', 'color2', 'color3', 'color4']].fillna("-")

# split data of column styles
def split_values(row):
    values = row['styles'].split(',')
    return pd.Series(values)

new_df = df.apply(split_values, axis=1)
new_df.columns = ['style{}'.format(i+1) for i in range(new_df.shape[1])]
df = pd.concat([df, new_df], axis=1)
df[['style1', 'style2', 'style3']] = df[['style1', 'style2', 'style3']].fillna("-")

# split data of column topicEffect
def split_values(row):
    values = row['topicEffect'].split(',')
    return pd.Series(values)

new_df = df.apply(split_values, axis=1)
new_df.columns = ['effect{}'.format(i+1) for i in range(new_df.shape[1])]
df = pd.concat([df, new_df], axis=1)
df[['effect1', 'effect2', 'effect3']] = df[['effect1', 'effect2', 'effect3']].fillna("-")


# Split data of column mood
def split_values(row):
    values = row['mood'].split(',')
    return pd.Series(values)

new_df = df.apply(split_values, axis=1)
new_df.columns = ['mood{}'.format(i+1) for i in range(new_df.shape[1])]
df = pd.concat([df, new_df], axis=1)
df[['mood1', 'mood2', 'mood3', 'mood4', 'mood5']] = df[['mood1', 'mood2', 'mood3', 'mood4', 'mood5']].fillna("-")

# Split data of column season
def split_values(row):
    values = row['season'].split(',')
    return pd.Series(values)

new_df = df.apply(split_values, axis=1)
new_df.columns = ['season{}'.format(i+1) for i in range(new_df.shape[1])]
nail_df = pd.concat([df, new_df], axis=1)
nail_df[['season1', 'season2']] = nail_df[['season1', 'season2']].fillna("-")

nail_df
```

```python
# Make a new subset with the variables that I need
subset_naildf = nail_df[["Image ID", "color1", "color2", "color3", "color4", "colorGroup", "shape", "length", "nailart", "nailartLevel", "style1", "style2", "style3", "effect1", "effect2", "effect3", "mood1", "mood2", "mood3", "mood4", "mood5", "season1", "season2"]]
subset_naildf
```

```python
# Presenting the questions that need to be asked to the user
# For this example I used the following input: color: light blue, shape: round, length: long, nail art: yes, level: simple, style: french tips, topic layer: shine, mood: confident, season: winter
inputColor = input("What color do you like to wear? (green, orange, blue, purple, yellow or red)")
inputColorGroup = input("Are you looking for a dark, light, pastel, neon, normal or mix" + inputColor + " ?")
inputShape = input("What shape do you like for your nails? (square, round or coffin)")
inputLength = input("What nail length do you prefer? (short, middle long or long)")
inputNailart = input("Do you want nail art on your nails? (yes/no)")
inputNailartLevel = input("What type of nailart would you like? (simple, medium or complex)")
inputNailartStyle = input("What nail art style do you like? (line art, swirl, patterns, dots, glitters, stones, small art, medium art, big art, french tips, reversed tip, hearts, flowers, checkered, stickers, ombre, marmer, golden flakes)")
inputTopicEffect = input("What would your topic layer be? (shine, shimmer, chrome, metallic or matt)")
inputMood = input("What mood would you like to be in? (calm, energetic, adventurous, vigorous, faithful, joyful, ambitious, proud, confident, powerful, masterful, trustworthy, happy, lively, loving, active, emotional)")
inputSeason = input("For what season are you designing your nails? (summer, spring, autumn or winter)")
```

```python
# Using the whole dataset as a dataframe to calculate the similarities of the user input on the variables.
# Inspiration code from Isha Bansal: https://www.askpython.com/python/examples/convert-number-to-words 

# Give score for nail shape
def like_color(input):
    nail_dataset = input
    
    if (nail_dataset == inputColor):
        return 1
    else:
        return 0
    
    
subset_naildf["score_color1"] = subset_naildf["color1"].apply(like_color)
subset_naildf["score_color2"] = subset_naildf["color2"].apply(like_color)
subset_naildf["score_color3"] = subset_naildf["color3"].apply(like_color)
subset_naildf["score_color4"] = subset_naildf["color4"].apply(like_color)
subset_naildf.head(5)
```

```python
#Give score for nail shape
def like_colorGroup(input):
    nail_dataset = input
    
    if (nail_dataset == inputColorGroup):
        return 1
    else:
        return 0
    
    
subset_naildf["score_colorGroup"] = subset_naildf["colorGroup"].apply(like_colorGroup)
subset_naildf.head(5)
```

```python
#Give score for nail shape
def like_shape(input):
    nail_dataset = input
    
    if (nail_dataset == inputShape):
        return 1
    else:
        return 0
    
    
subset_naildf["score_shape"] = subset_naildf["shape"].apply(like_shape)
subset_naildf.head(5)
```

```python
# Give score for length of the nail

def like_length(input):
    nail_dataset = input
    
    if (nail_dataset == inputLength):
        return 1
    else:
        return 0   
    
subset_naildf["score_length"] = subset_naildf["length"].apply(like_length)
subset_naildf.head(5)
```

```python
# Give score for the use of nail art

def use_nailart(input):
    nail_dataset = input
    if (nail_dataset == inputNailart):
        return 1
    else:
        return 0   
    
subset_naildf["score_nailartUse"] = subset_naildf["nailart"].apply(use_nailart)
subset_naildf.head(5)
```

```python
# Give score for the complexity level of the nail art
def nailart_level(input):
    nail_dataset = input
    if (nail_dataset == inputNailartLevel):
        return 1
    else:
        return 0   
    
subset_naildf["score_nailartLevel"] = subset_naildf["nailartLevel"].apply(nailart_level)
subset_naildf.head(5)
```

```python
# Give score for nailart style

def like_nailart(input):
    nail_dataset = input
    if (nail_dataset == inputNailartStyle):
        return 1
    else:
        return 0   
    
subset_naildf["score_style1"] = subset_naildf["style1"].apply(like_nailart)
subset_naildf["score_style2"] = subset_naildf["style2"].apply(like_nailart)
subset_naildf["score_style3"] = subset_naildf["style3"].apply(like_nailart)
subset_naildf
```

```python
# Give score for topic effect

def like_topicEffect(input):
    nail_dataset = input
    if (nail_dataset == inputTopicEffect):
        return 1
    else:
        return 0   
    
subset_naildf["score_effect1"] = subset_naildf["effect1"].apply(like_topicEffect)
subset_naildf["score_effect2"] = subset_naildf["effect2"].apply(like_topicEffect)
subset_naildf["score_effect3"] = subset_naildf["effect3"].apply(like_topicEffect)
subset_naildf
```

```python
# Give score for mood

def like_mood(input):
    nail_dataset = input
    if (nail_dataset == inputMood):
        return 1
    else:
        return 0   
    
subset_naildf["score_mood1"] = subset_naildf["mood1"].apply(like_mood)
subset_naildf["score_mood2"] = subset_naildf["mood2"].apply(like_mood)
subset_naildf["score_mood3"] = subset_naildf["mood3"].apply(like_mood)
subset_naildf["score_mood4"] = subset_naildf["mood4"].apply(like_mood)
subset_naildf["score_mood5"] = subset_naildf["mood5"].apply(like_mood)
subset_naildf
```

```python
# Give score for season

def like_season(input):
    nail_dataset = input
    if (nail_dataset == inputSeason):
        return 1
    else:
        return 0   
    
subset_naildf["score_season1"] = subset_naildf["season1"].apply(like_mood)
subset_naildf["score_season2"] = subset_naildf["season2"].apply(like_mood)
subset_naildf
```

```python
# Assigning weight to the different variables, so the end score will be more accurate.
weights = {'var1': 0.2, 'var2': 0.3, 'var3': 0.5}

subset_naildf["Final_score"] = ((subset_naildf["score_color1"] * weights['var3']) + (subset_naildf["score_color2"] * weights['var3']) + (subset_naildf["score_color3"] * weights['var3']) + (subset_naildf["score_color4"] * weights['var3']) + (subset_naildf["score_colorGroup"] * weights['var3']) + (subset_naildf["score_shape"] * weights['var1']) + (subset_naildf["score_length"] * weights['var2']) + (subset_naildf["score_nailartUse"] * weights['var3']) + (subset_naildf["score_nailartLevel"] * weights['var3']) + (subset_naildf["score_style1"] * weights['var3']) + (subset_naildf["score_style2"] * weights['var3']) + (subset_naildf["score_style3"] * weights['var3']) + (subset_naildf["score_effect1"] * weights['var2']) + (subset_naildf["score_effect2"] * weights['var2']) + (subset_naildf["score_effect3"] * weights['var2']) + (subset_naildf["score_mood1"] * weights['var3']) + (subset_naildf["score_mood2"] * weights['var3']) + (subset_naildf["score_mood3"] * weights['var3'])+ (subset_naildf["score_mood4"] * weights['var3']) + (subset_naildf["score_mood5"] * weights['var3']) + (subset_naildf["score_season1"] * weights['var2']) + (subset_naildf["score_season2"] * weights['var2']))
subset_naildf["Final_score"] = (subset_naildf["Final_score"] / 10) * 100
subset_naildf = subset_naildf.sort_values(by="Final_score", ascending=False)
subset_naildf
```

```python
# Making a new subset with only the scores of the variables
subset_final_scores = subset_naildf[["Image ID", "score_color1", "score_color2", "score_color3", "score_color4", "score_colorGroup", "score_shape", "score_length", "score_nailartUse", "score_nailartLevel", "score_style1", "score_style2", "score_style3", "score_effect1", "score_effect2", "score_effect3", "score_mood1", "score_mood2", "score_mood3", "score_mood4", "score_mood5", "score_season1", "score_season2", "Final_score"]]
subset_final_scores
```

```python
# Showing the final score of the 5 best recommendations
final_image_score = subset_final_scores[["Image ID", "Final_score"]]
final_image_score.head(5)
```
